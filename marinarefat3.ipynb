```python
import cobra
```

```python
from cobra import Model ,Reaction , Metabolite
```

```python

model = Model("Glucose_Reaction")
```

```python
###################  Glu+ATP ==============>> G6P+ADP+H  ***** HK
HK=Reaction("HK")
HK.name="Hexokinase"
HK.lower_bound=0
HK.upper_bound=1000

###################  G6P  <<==============>> F6P ******* PGI
PGI=Reaction("PGI")
PGI.name="Glucose 6-phosphate isomerase"
PGI.lower_bound=-1000
PGI.upper_bound=1000


##################   F6P + ATP ==============>> FDP+ADP+H***** PFK
PFK=Reaction("PFK")
PFK.name="Phosphofructokinase"
PFK.lower_bound=0
PFK.upper_bound=1000

  
    
################  DHAP <<==============>> GAP****** TPI   
TPI=Reaction("TPI")
TPI.name="Triose-phosphate isomerase"
TPI.lower_bound=-1000
TPI.upper_bound=1000    

################  FDP <<==============>> GAP+DHAP***** ALD
ALD=Reaction("ALD")
ALD.name="Fructose 1,6-diphosphate aldolase"
ALD.lower_bound=-1000
ALD.upper_bound=1000



################  GAP + NAD + Pi <<==============>> DPG13 + NADH + H****  GAPDH 
GAPDH=Reaction("GAPDH")
GAPDH.name="Glyceraldehyde 3-phosphate dehydrogenase"
GAPDH.lower_bound=-1000
GAPDH.upper_bound=1000

################   DPG13 + ADP <<==============>>  PG3 + ATP*****   PGK    
PGK=Reaction("PGK")
PGK.name="Phosphoglycerate kinase"
PGK.lower_bound=-1000
PGK.upper_bound=1000


################   PG3<<==============>>  PG2******  PGLM     
PGLM=Reaction("PGLM")
PGLM.name="Phosphoglycerate mutase"
PGLM.lower_bound=-1000
PGLM.upper_bound=1000


################    PG2 <<==============>>  PEP + H2O*****   ENO      
ENO=Reaction("ENO")
ENO.name="Enolase"
ENO.lower_bound=-1000
ENO.upper_bound=1000

################   PEP + ADP + H+ <<==============>>  PYR + ATP*****  PK     
PK=Reaction("PK")
PK.name="Pyruvate kinase"
PK.lower_bound=0
PK.upper_bound=1000

################    ==============>>AMP  ******   AMPenter    
AMPenter=Reaction("AMPenter")
AMPenter.name="AMP export"
AMPenter.lower_bound=0
AMPenter.upper_bound=1000


################  2ADP <<<==============>>AMP + ATP ***** APK  

APK=Reaction("APK")
APK.name="Adenylate kinase"
APK.lower_bound=-1000
APK.upper_bound=1000


################  PYR <<<==============>> ***** PYR 

PYR=Reaction("PYR")
PYR.name="Pyruvate exchange"
PYR.lower_bound=-1000
PYR.upper_bound=1000


################   ATP + H2O ==============>>  ADP + Pi + H+*****ATP   

ATP=Reaction("ATP")
ATP.name="ATP hydrolysis"
ATP.lower_bound=0
ATP.upper_bound=1000


################  NADH ==============>>NAD + H+ ****** NADH

NADH=Reaction("NADH")
NADH.name="NADH oxidation"
NADH.lower_bound=0
NADH.upper_bound=1000
                   
    
    
    
####################   ==============>>Glc*****  GLUin 
GLUin=Reaction("GLUin")
GLUin.name="Glucose import"
GLUin.lower_bound=1
GLUin.upper_bound=1    



################   AMP  ==============>> **** AMPout

AMPout=Reaction("AMPout")
AMPout.name="AMP export"
AMPout.lower_bound=0
AMPout.upper_bound=1000

################    H <<==============>>  ***** H 

H=Reaction("H")
H.name="Proton exchange"
H.lower_bound=-1000
H.upper_bound=1000

################    H2o <<==============>>  ******H2o

H2o=Reaction("H2o")
H2o.name="Water exchange"
H2o.lower_bound=-1000
H2o.upper_bound=1000


    


    
    







    
```

```python

Glc=Metabolite("Glc",
                 formula="C6H12O6",
                 name="Glucose",
                 compartment="c")

G6P=Metabolite("G6P",
                 formula="G6H13O9P",
                 name="Glucose 6-phosphate",
                 compartment="c")

F6P=Metabolite("F6P",
                 formula="C6H13O9P",
                 name="Fructose 6-phosphate",
                 compartment="c")

FDP=Metabolite("FDP",                         
                 formula="C6H13O9P",
                 name="Fructose 1,6-diphosphate",
                 compartment="c")




DHAP=Metabolite("DHAP",
                 formula="C3H7O6P",
                 name="Dihydroxyacetone phosphate",
                 compartment="c")


GAP=Metabolite("GAP",
                 formula="C3H7O6P",
                 name="Glyceraldehyde 3-phosphate",
                 compartment="c")


DPG13=Metabolite("DPG13",              
                 formula="C3H6O10P2",
                 name="1,3-Diphosphoglycerate",
                 compartment="c")




PG3=Metabolite("PG3",
                 formula="C3H7O7P",
                 name="3-Phosphoglycerate",
                 compartment="c")
PG2=Metabolite("PG2",
                 formula="C3H7O7P",
                 name="2-Phosphoglycerate",
                 compartment="c")
PEP=Metabolite("PEP",
                 formula="C3H5O6P",
                 name="Phosphoenolpyruvate",
                 compartment="c")

PYRm=Metabolite("PYRm",
                 formula="C3H4O3",
                 name="Pyruvate",
                 compartment="c")

NAD=Metabolite("NAD",
                 formula="C21H27N7O14P2",
                 name="Nicotinamide adenine dinucleotide oxidized",
                 compartment="c")

NADHm=Metabolite("NADHm",
                 formula="C21H27N7O14P2",
                 name="Nicotinamide adenine dinucleotide reduced",
                 compartment="c")


AMPm=Metabolite("AMPm",
                 formula="C3H4O3",
                 name="Adenosine mono-phosphate",
                 compartment="c")


ADP=Metabolite("ADP",
                 formula="C10H15N5O10P2",  
                 name="Adenosine di-phosphate",
                 compartment="c")
ATPm=Metabolite("ATPm",
                 formula="C10H16N5O13P3",
                 name="Adenosine tri-phosphate",
                 compartment="c")

Pi =Metabolite("Pi",
                 formula="PO4",
                 name="norganic phosphate",        
                 compartment="c")



H2Om=Metabolite("H2Om",
                 formula="H2O",
                 name="Water",
                 compartment="c")


Hm  =Metabolite("Hm", 
                 formula="H",
                 name="Proton",
                 compartment="c")

```

```python

GLUin.add_metabolites({Glc:1}) 
HK.add_metabolites({Glc:-1,ATPm:-1,G6P:1,ADP:1,Hm:1})
PGI.add_metabolites({G6P:-1,F6P:1})
PFK.add_metabolites({F6P:-1,ATPm:-1,FDP:1,ADP:1,Hm:1})
ALD.add_metabolites({FDP:-1,GAP:1,DHAP:1})
TPI.add_metabolites({DHAP:-1,GAP:1})
GAPDH.add_metabolites({GAP:-1,NAD:-1,Pi:-1  ,DPG13:1,NADHm :1,Hm:1})
PGK.add_metabolites({DPG13:-1 ,ADP:-1,PG3:1,ATPm:1})  
PGLM.add_metabolites({PG3:-1,PG2:1})   
ENO.add_metabolites({PG2:-1,PEP:1, H2Om:1}) 
PK.add_metabolites({PEP:-1,ADP:-1,Hm:-1,PYRm:1,ATPm:1})  
AMPenter.add_metabolites({AMPm:1})  
H.add_metabolites({Hm:-1})
H2o.add_metabolites({H2Om:-1})
ATP.add_metabolites({ ATPm:-1, H2Om:-1, ADP:1 ,Pi:1,Hm:1})   
AMPout.add_metabolites({AMPm:-1})   
APK.add_metabolites({ADP:-2,AMPm:1,ATPm:1})  
PYR.add_metabolites({PYRm:-1})     
NADH.add_metabolites({NADHm:-1,NAD:1,Hm:1})    
     
```

```python
model.add_reactions([HK,PGI,PFK,ALD,TPI,GAPDH,PGK,PGLM,ENO,APK,PK,ATP,AMPout,PYR,NADH,GLUin,AMPenter,H,H2o])
```

```python
model.objective='PK'
```

```python
model.optimize()
```

```python
cobra.io.save_json_model(model,"Glucose_Reaction.json")
```
